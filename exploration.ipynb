# Fashion-MNIST Exploration

Trying different architectures and hyperparameters.

```python
import torch
import torchvision
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import numpy as np
```

```python
transform = transforms.Compose([transforms.ToTensor()])
train = datasets.FashionMNIST('data', train=True, download=True, transform=transform)
test = datasets.FashionMNIST('data', train=False, transform=transform)
print(f'Train: {len(train)}, Test: {len(test)}')
```

```python
# visualize some samples
fig, axes = plt.subplots(2, 5, figsize=(12, 5))
for i, ax in enumerate(axes.flat):
    img, label = train[i]
    ax.imshow(img.squeeze(), cmap='gray')
    ax.set_title(train.classes[label])
    ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
# tried basic CNN first - got ~93.4%
# then resnet with SE blocks - 94.7%
# mixup didn't help much, maybe 0.1-0.2% improvement
# label smoothing=0.1 was good though
# cutout was helpful on the CNN but less so on resnet
```
